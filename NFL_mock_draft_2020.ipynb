```python
import pandas as pd

mock_drafts = pd.read_excel(r"C:\Users\Jonathan\nfl_mock_drafts_2020.xlsx")
```

```python
#remove everything after , (year and trade info)
for i in mock_drafts.columns:
    mock_drafts[i] = mock_drafts[i].str.split(',').str[0]
```

```python
import re
#remove everything after second capitalized word
for i in mock_drafts.columns:
    first = mock_drafts[i].str.split(' ').str[0]
    t=mock_drafts[i].str.split(' ').str[1]
    result = []
    for j in range(0, len(t)):
        last = re.findall('[A-Z][^A-Z]*', t[j])[0]
        result.append(first[j]+ ' ' + last)
    mock_drafts[i] = result
```

```python
#remove non character values 
for i in mock_drafts.columns:
    mock_drafts[i] = mock_drafts[i].str.replace('[^a-zA-Z ]', '')
```

```python
#create dictionary of players and the number of the 16 mock drafts they appear in 
players_dict = dict()
for i in mock_drafts.columns:
    col = mock_drafts[i]
    for j in col:
        if j in players_dict.keys():
            players_dict[j] +=1
        else:
            players_dict[j] = 1

sorted_players_dict = sorted(players_dict.items(), key=lambda x: x[1])
sorted_players_dict
```

```python
#fix data quality issues
for i in mock_drafts.columns:
    mock_drafts[i] = mock_drafts[i].str.replace('Yetur Gross', 'Yetur Gross-Matos')
    mock_drafts[i] = mock_drafts[i].str.replace('Joshua Jones', 'Josh Jones')
    mock_drafts[i] = mock_drafts[i].str.replace('Tua Tagavailoa', 'Tua Tagovailoa')
    mock_drafts[i] = mock_drafts[i].str.replace('Xavier Mc', 'Xavier McKinney')
    mock_drafts[i] = mock_drafts[i].str.replace('Justin Herbery', 'Justin Herbert')
    mock_drafts[i] = mock_drafts[i].str.replace('Kenneth Murrray', 'Kenneth Murray')
    mock_drafts[i] = mock_drafts[i].str.replace('Isaiah Wlson', 'Isaiah Wilson')
    mock_drafts[i] = mock_drafts[i].str.replace('Yeter Gross', 'Yetur Gross-Matos')
    mock_drafts[i] = mock_drafts[i].str.replace('Jaylon Reagor', 'Jalen Reagor')
    mock_drafts[i] = mock_drafts[i].str.replace('Jery Jeudy', 'Jerry Jeudy')
    mock_drafts[i] = mock_drafts[i].str.replace('Jalon Reagor', 'Jalen Reagor')
    mock_drafts[i] = mock_drafts[i].str.replace('Joe Burrw', 'Joe Burrow')
    mock_drafts[i] = mock_drafts[i].str.replace('Grand Delpit', 'Grant Delpit')
    
#recreate dictionary of players and the number of the 16 mock drafts they appear in to see if we need to fix anything else
players_dict = dict()
for i in mock_drafts.columns:
    col = mock_drafts[i]
    for j in col:
        if j in players_dict.keys():
            players_dict[j] +=1
        else:
            players_dict[j] = 1

sorted_players_dict = sorted(players_dict.items(), key=lambda x: x[1])
sorted_players_dict
```

```python
#accumulate the rank positions of each player across all mock drafts
rank_position_dict = dict()
for i in mock_drafts.columns:
    col = mock_drafts[i]
    for j in range(0, len(col)):
        if col[j] in rank_position_dict.keys():
            rank_position_dict[col[j]].append(j+1)
        else:
            rank_position_dict[col[j]] = [j+1]
```

```python
import numpy as np
from functools import reduce
from scipy.stats import mode
#accumulate mean of each rec set, variance of each rec set, length of rec set
player_ranks = pd.DataFrame()
add = 33
#add = 64
#add = 48
for i in rank_position_dict.keys():
    player_info = []
    rec_set = rank_position_dict[i]
    #add prod rank, mean, variance, len, min and max to player_info
    player_info.append(reduce(lambda x, y: x * y, rec_set, 1)**(1/len(rec_set)))
    player_info.append(np.mean(rec_set))
    player_info.append(np.var(rec_set))
    player_info.append(len(rec_set))
    player_info.append(min(rec_set))
    player_info.append(max(rec_set))
    player_info.append(mode(rec_set)[0][0])
    #if a player is not in all mock drafts
    #then add 33 n times to their rec set where n = # mock drafts they are not in
    #SHOULD WE ADD 33 OR SOMETHING LARGER TO PENALIZE FOR NOT BEING IN ALL MOCK DRAFTS??
    if len(rec_set) != len(mock_drafts.columns):
        remaining = [add]*(len(mock_drafts.columns)-len(rec_set))
        rec_set.extend(remaining)
    #add new prod rank, mean, variance, length, min and max to player_info
    player_info.append(reduce(lambda x, y: x * y, rec_set, 1)**(1/len(rec_set)))
    player_info.append(np.mean(rec_set))
    player_info.append(np.var(rec_set))
    #should be the same for all players, the number of mock drafts we use 
    player_info.append(len(rec_set))
    player_info.append(min(rec_set))
    player_info.append(max(rec_set))
    player_ranks[i]=player_info

player_ranks.index = ['Original Rank Product', 'Original Mean', 'Original Variance', 'Original Length', 'Original Min', 'Original Max', 'Original Mode', 'New Rank Product', 'New Mean', 'New Variance', 'New Length', 'New Min', 'New Max']
player_ranks
```

```python
#rank players based on original mean and new mean
player_ranks = player_ranks.T
player_ranks = player_ranks[player_ranks['Original Length']>=25]
player_ranks['Original Mean Rank'] = player_ranks['Original Mean'].rank()
player_ranks['New Mean Rank'] = player_ranks['New Mean'].rank()
player_ranks['Original Rank Product Rank'] = player_ranks['Original Rank Product'].rank()
player_ranks['New Rank Product Rank'] = player_ranks['New Rank Product'].rank()
player_ranks.sort_values(by='New Rank Product Rank')
```
