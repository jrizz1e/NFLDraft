```python
#read in our picks and the actual draft order
import pandas as pd

picks = pd.read_excel('picks.xlsx')
picks.index = list(range(1,33))

actual_draft_picks = pd.read_excel('actual_picks.xlsx')
actual_draft_picks.index = list(range(1,107))

picks
```

```python
actual_draft_picks
```

```python
players = picks.apply(pd.Series.value_counts)
players['selected by'] = players.sum(axis=1).astype(int)
players.sort_values(by=['selected by'])
```

```python
#real time scoring
from math import exp

users = []
pick_differences = []
spot_ons = []
remainings = []
for i in picks.columns:
    pick_abs_difference = 0
    spot_on = 0
    remaining = 0 
    for j in picks[i]:
        user_pick = picks.index[picks[i] == j].tolist()[0]
        actual_pick = actual_draft_picks.index[actual_draft_picks['Picks']==j].tolist()
        #print(j, user_pick, actual_pick)
        if len(actual_pick) > 0:
            actual_pick = actual_pick[0]
            if user_pick == actual_pick:
                spot_on += 1
            pick_abs_difference += abs(actual_pick-user_pick)
        else:
            remaining+=1
            print(i, j)
    #print(i, pick_abs_difference, spot_on, remaining)
    users.append(i)
    pick_differences.append(pick_abs_difference)
    spot_ons.append(spot_on)
    remainings.append(remaining)
    
results = pd.DataFrame()
results['Users']=users
results['Total Pick Deviation'] = pick_differences
results['Number of Spot on Picks'] = spot_ons
results['Number of Players Left'] = remainings
results.sort_values(by=['Total Pick Deviation'])
```
